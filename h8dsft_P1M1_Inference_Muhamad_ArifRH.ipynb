# 1- Import Library

```python
import pickle
import json
import pandas as pd
import numpy as np
```

# 2 - Load Models and Data Sample

```python
# Load Files
with open('svc_tuned.pkl', 'rb') as file_1:
    model = pickle.load(file_1)

with open('selected_col.txt', 'r') as file_4:
    selected_col = json.load(file_4)

with open('data_sample.txt', 'r') as file_6:
    data_sample = json.load(file_6)

# Convert JSON file to Pandas Dataframe
pd.set_option('display.max_columns', None)
data_inference = pd.read_json(data_sample)
data_inference
```

# 3 - Model Inferencing

```python
# Display Selected Feature
selected_col
```

```python
# Selected Features
data_inference_sel = data_inference[selected_col]
```

```python
# Predict Income Bracket
y_pred_inferencing = model.predict(data_inference_sel)
y_pred_inferencing
```

```python
# Showing reference dataframe
data_inference['Prediction Results'] = y_pred_inferencing.copy()
data_inference
```
